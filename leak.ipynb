```python
from diffeqpy import de


def F(du, u, params, t):
    
    resid = []
    resid.append(du[0] + u[0])
    
    
    return resid

u0 = [1.0]
du0 = [-1.0]

tspan = (0.0, 1.0)
differential_vars = [True]

prob = de.DAEProblem(F, du0, u0, tspan, differential_vars=differential_vars)
sol = de.solve(prob)
```

```python
import pylab as plt


y_data = [y[0] for y in sol.u]
plt.plot(sol.t, y_data)

plt.show()
```
