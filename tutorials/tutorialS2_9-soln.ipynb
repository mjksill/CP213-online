## Model Solutions for Question 1

The set of differential equations that need to be solved can be written 
in matrix form as
$$\begin{align*}
    \frac{d}{dt} \vec{x} &= {\bf A}\cdot\vec{x}
    .
\end{align*}$$
To solve the problem we must decouple the
differential equations by performing a transformation so 
that the matrix $A$ is diagonal.  The matrix of
eigenvectors ${\bf P}$, where the columns of ${\bf P} = \begin{pmatrix} 2 & -1 \\ 1 & 1 \end{pmatrix}$ are the eigenvectors
of the matrix ${\bf A}$ which transforms the system to the eigenbasis.
As in class, we began by introducing the new function ${\bf P}\vec{u}=\vec{x}$ which when differentiated becomes:
$$\begin{align*}
    {\bf P}\cdot\frac{d}{dt}\vec{u}=\frac{ d }{ dt }\vec{x}
\end{align*}$$
Substiuting these terms into our original equation
$$\begin{align*}
    {\bf P}\cdot\frac{d}{dt} \vec{u} 
    &= {\bf A}\cdot{\bf P}\cdot\vec{u}
\end{align*}$$
Multiplying both sides of the equation by ${\bf P}^{-1}$
$$\begin{align*}
    \frac{d}{dt} \vec{u} 
    &= {\bf P}^{-1}\cdot{\bf A}\cdot{\bf P}\cdot\vec{u}
\end{align*}$$

We note that the matrix ${\bf D}={\bf P}^{-1} {\bf A}\cdot{\bf P}$ is diagonal, and its elements are the eigenvalues of the matrix ${\bf A}$, hence ${\bf D} = \begin{pmatrix} 0 & 0 \\ 0 & -0.75 \end{pmatrix}$. This can now be rewritten as
$$\begin{align*}
    \frac{d}{dt}\vec{u} 
    &= {\bf D}\cdot\vec{u}
    .
\end{align*}$$

This matrix equation can be written in component form as:
$$\begin{align*}
    \frac{du_k}{dt}
    &= \lambda_k u_k
\end{align*}$$
where $\lambda_k$ is the $k\text{th}$  eigenvalue of the matrix ${\bf A}$.
Each of these differential equations can be solved using separation of variables:
$$\begin{align*}
    \int_{u_k(0)}^{u_k(t)}\frac{du_k}{u_k}
    &= \int_{0}^{t}\, \lambda_k\,dt
    \\
    \ln{\frac{u_k}{u_k(0)}}
    &= \lambda_k t
    \\
    u_k(t)
    &= u_k(0)e^{\lambda_k t}
\end{align*}$$

Hence, for our two tank system we have  decoupled solutions as

$$\begin{align*}
    u_1(t)
    &= u_1(0)e^{0 t}
    \\
    &= u_1(0)
    \\
    u_2(t)
    &=u_2(0)e^{-0.75t}
\end{align*}$$


We can determine $\vec{u}(0)$ from

$$\begin{align*}
    \vec{u}(0)&={\bf P}^{-1}\cdot\vec{x}(0)
\end{align*}$$

Where ${\bf P}^{-1} = \frac{1}{3}\begin{pmatrix} 1 & 1\\ -1 & 2 \end{pmatrix}$

$$\begin{align*}
    \begin{pmatrix} u_1(0) \\ u_2(0) \end{pmatrix}&=\frac{1}{3}\begin{pmatrix} 1 & 1\\ -1 & 2 \end{pmatrix}\cdot\begin{pmatrix} 5 \\ 4 \end{pmatrix}
    \\
    \begin{pmatrix} u_1(0) \\ u_2(0) \end{pmatrix}&=\begin{pmatrix} 3 \\ 1 \end{pmatrix}
\end{align*}$$

The solution in terms of $\vec{u}(t)$ is given by
$$\begin{align*}
    \vec{u}(t)
    &= 
    \begin{pmatrix} 3 \\ e^{-0.75t} \end{pmatrix}
\end{align*}$$

The mass of salt in tank $k$, ${\bf x_k(t)}$ and can be determined from $\vec{x}(t)={\bf P}\cdot\vec{u}(t)$.

$$\begin{align*}
    \begin{pmatrix} x_1(t) \\ x_2(t)\end{pmatrix}&=\begin{pmatrix} 2 & -1\\ 1 & 1 \end{pmatrix}\cdot\begin{pmatrix} 3 \\ e^{-0.75t} \end{pmatrix}
\end{align*}$$

Hence the mass of the salt in each tank is given by

$$\begin{align*}
    x_1(t) &= 6 - e^{-0.75t}
    \\
    x_2(t) &= 3 + e^{-0.75t}
\end{align*}$$

Notice what happens to our system as $t\rightarrow \infty$, the mass of the tanks become

$$\begin{align*}
    x_1(\infty)=6 \ \rm{kg}
    \\
    x_2(\infty)=3\ \rm{kg}
\end{align*}$$

Which makes sense since we would anticipate that concentration in the full system to reach an equillibrium value equivelent to the total mass in the system ($x_1(0)+x_2(0)$) per total volume in the system ($V_1 +V_2$).

```python
import numpy as np
from numpy.linalg import eig, inv


V1 = 40.0  # volume of tank 1 / L
m1 = 5.0   # initial mass of salt in tank 1 / kg
V2 = 20.0  # volume of tank 2 / L
m2 = 4.0   # initial mass of salt in tank 2 / kg

dotQ = 10.0 # recirculation rate / L min^{-1}

xinit = np.array([m1, m2])
print('initial composition vector')
print(xinit)

# setup A matrix
A = np.array([[-dotQ/V1,  dotQ/V2],
              [ dotQ/V1, -dotQ/V2]])
print()
print('A matrix')
print(A)

# compute eigenvalue and eigenvectors
lam, P = eig(A)
print()
print('eigenvalues')
print(lam)

print()
print('eigenvectors')
print(P)

# obtain initial conditions in terms of u
uinit = np.dot(inv(P), xinit)
print()
print('u(0) = ')
print(uinit)


t_data = np.arange(0.0, 10.0, 0.01)

x1_data = []
x2_data = []
for t in t_data:
    u1 = uinit[0]*np.exp(lam[0]*t) # note that Python starts counting at 0
    u2 = uinit[1]*np.exp(lam[1]*t)
    ut = np.array([u1, u2])
    
    xt = np.dot(P, ut)
    x1_data.append( xt[0] )
    x2_data.append( xt[1] )

    
import pylab as plt
plt.rcParams['font.size'] = 18
plt.rcParams['axes.labelsize'] = 'large'
plt.rcParams.update({'figure.autolayout': True})

plt.plot(t_data, x1_data, label='tank 1')
plt.plot(t_data, x2_data, label='tank 2')
plt.xlabel('time / min')
plt.ylabel(r'mass of salt / kg')
plt.legend()
plt.show()
```

## Model solutions for Q2

### Part 1


The set of differential equations that need to be solved can be written 
in matrix form as
$$\begin{align*}
\frac{d}{dt} \vec{T} &= {\bf A}\cdot\vec{T} + \vec{b}
.
\end{align*}$$
The basic strategy to solve the problem is to decouple the
differential equations from each other by performing a coordinate
transformation so that the matrix $A$ is diagonal.  The matrix of
eigenvectors ${\bf P}$, where the columns of $P$ are the eigenvectors
of the matrix ${\bf A}$ provides the required transformation.
Multiplying both sides of the equation by ${\bf P}^{-1}$
$$\begin{align*}
{\bf P}^{-1}\cdot\frac{d}{dt} \vec{T} 
&= {\bf P}^{-1} \cdot{\bf A}\cdot\vec{T} + {\bf P}^{-1}\cdot\vec{b}
\\
\frac{d}{dt} ({\bf P}^{-1}\cdot\vec{T} )
&= {\bf P}^{-1} {\bf A}\cdot{\bf P}\cdot{\bf P}^{-1}
\cdot\vec{T} + {\bf P}^{-1}\cdot\vec{b}
.
\end{align*}$$
We note that the matrix ${\bf D}={\bf P}^{-1} {\bf A}\cdot{\bf P}$ is diagonal, and its elements are the eigenvalues of the matrix ${\bf A}$.  If we define a new set of variables $\vec{y}={\bf P}^{-1}\cdot\vec{T}$, which is simply a translation of the temperature vector, the matrix equation can be rewritten as
$$\begin{align*}
\frac{d}{dt} \vec{y} 
&= {\bf D}\cdot\vec{y} + \vec{b}'
.
\end{align*}$$

This matrix equation can be written in component form as:
$$\begin{align*}
\frac{dy_k}{dt}
&= \lambda_k y_k + b_k'
\end{align*}$$
where $\lambda_k$ is the $k{\rm th}$ eigenvalue of the matrix ${\bf A}$.
Each of these differential equations can be solved using separation of variables:
$$\begin{align*}
\frac{dy_k}{\lambda_k y_k + b_k'}
&= dt
\\
\frac{1}{\lambda_k}\ln \frac{\lambda_k y_k + b_k'}{\lambda_k y_k(0) + b_k'}
&= t
\\
\lambda_k y_k + b_k'
&= (\lambda_k y_k(0) + b_k')e^{\lambda_k t}
\\
y_k(t)
&= y_k(0)e^{\lambda_k t}
- \frac{b_k'}{\lambda_k} (1-e^{\lambda_k t})
\end{align*}$$

The temperatures can be determined from the variables $y$ from $\vec{T}(t)={\bf P}\cdot\vec{y}(t)$.


```python
import numpy as np
from numpy.linalg import eig, inv

N = 3  # note that N must be greater than 2
T0 = np.array([10.0, 80.0, 10.0]).reshape(N,1)
t_data = np.arange(0.0, 100.0, 0.01)

dotVRin = 0.1
dotVLout = 0
dotVLin = 0
dotVRout = 0.1
dotvR = 0.2
dotvL = 0.1
V = 1.0

TRin = 80.0
TLin = 10.0

# construct the matrix A
A = np.zeros((N,N))
A[0,0] = -(dotvR+dotVLout)/V
A[0,1] = dotvL/V
A[N-1,N-2] = dotvR/V
A[N-1,N-1] = -(dotvL+dotVRout)/V
for k in range(1,N-1):
    A[k,k-1] = dotvR/V
    A[k,k] = -(dotvL+dotvR)/V
    A[k,k+1] = dotvL/V
print('A matrix')
print(A)

# construct the b vector
b = np.zeros((N,1))
b[0,0] = TRin*dotVRin/V
b[N-1,0] = TLin*dotVLin/V
print(r'b vector')
print(f'{b}')



lam, P = eig(A)
print()
print('eigenvalues')
print(lam)

print()
print('eigenvectors')
print(P)

y0 = np.dot(inv(P), T0)

bp = np.dot(inv(P), b)
print()
print('bp = ')
print(bp)


y_data = []
for t in t_data:
    yt = []
    for k in range(N):
        lamk = lam[k]
        bpk = bp[k,0]
        yt.append( y0[k,0]*np.exp(lamk*t) - (1.0-np.exp(lamk*t))*bpk/lamk )
    y_data.append(yt)

T_data = [[] for k in range(N)]
for y in y_data:
    y_vec = np.array(y).reshape(N,1)
    T_vec = np.dot(P, y_vec)
    for k in range(N):
        T_data[k].append(T_vec[k,0])

import pylab as plt
plt.rcParams['font.size'] = 18
plt.rcParams['axes.labelsize'] = 'large'
plt.rcParams.update({'figure.autolayout': True})

for k in range(N):
    plt.plot(t_data, T_data[k], label='tank '+str(k))
plt.xlabel(r'time / seconds')
plt.ylabel(r'temperature / $^\circ$C')
plt.legend()
plt.show()

```
