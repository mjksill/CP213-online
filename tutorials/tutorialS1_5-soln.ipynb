## Model solutions for Q1

The downward pressure force of the water on a small section of the
block of height $dy$ located at a height $y$ from the bottom of the
block is given by
\begin{align*}
dF_y &= - p(y) \frac{W dy}{\sin\theta} \cos\theta
\\
&= - [p_0 + \rho g(H-y)] \frac{W dy}{\sin\theta} \cos\theta 
.
\end{align*}
By integrating this force and including the force due to the
atmospheric pressure, we find that the total pressure force in the
$y$-direction is
\begin{align*}
F_y &= - \int_0^H [p_0 + \rho g(H-y)] \frac{W dy}{\sin\theta} \cos\theta
- p_0 (L-H\cot\theta)W
\\
&= - \left[(p_0 + \rho gH)y-\rho g\frac{y^2}{2}\right]_0^H W\cot\theta
- p_0 (L-H\cot\theta)W
\\
&= - \left[(p_0 + \rho gH)H-\rho g\frac{H^2}{2}\right] W\cot\theta
- p_0 (L-H\cot\theta)W
\\
&= - \left[p_0 + \rho g\frac{H}{2}\right] WH \cot\theta
- p_0 (L-H\cot\theta)W
\\
&= - \rho g\frac{WH^2}{2}\cot\theta
- p_0 LW
.
\end{align*}

The force in the $x$-direction due to the pressure on a small section
of the block of height $dy$ located at a height $y$ from the bottom of
the block is given by
\begin{align*}
dF_x &= p(y) \frac{W dy}{\sin\theta} \sin\theta - p_0 Wdy
\\
&= [p_0 + \rho g(H-y)] \frac{W dy}{\sin\theta} \sin\theta  - p_0 Wdy
\\
&= [p_0 + \rho g(H-y)] W dy  - p_0 Wdy
.
\end{align*}
By integrating this force across the entire surface of the block, we
find the net force in the $x$-direction is
\begin{align*}
F_x &= \int_0^H [p_0 + \rho g(H-y)] W dy  - \int_0^H p_0 Wdy
\\
&= \left[(p_0 + \rho g H)y -\rho g \frac{y^2}{2}) - p_0 \right]_0^H W
\\
&= \left[(p_0 + \rho g H)H -\rho g \frac{H^2}{2}) - p_0H \right] W
\\
&= \rho g \frac{WH^2}{2}
.
\end{align*}

The block will move when the net pressure force pushing the block to
the right is becomes greater than the friction force:
\begin{align*}
\rho g \frac{WH^2}{2} 
&= \mu\left[mg + \rho g\frac{WH^2}{2}\cot\theta + p_0 LW\right]
\\
\rho g \frac{H^2}{2} 
&= \mu\left[\frac{mg}{W} + \rho g\frac{H^2}{2}\cot\theta + p_0 L\right]
\\
\rho g \frac{H^2}{2}(1-\mu \cot\theta) &= \mu \left(\frac{mg}{W}+p_0 L\right)
\\
H &= \left[\frac{2\mu (mg/W+p_0 L)}{\rho g (1-\mu \cot\theta)}\right]^{1/2}
.
\end{align*}


## Question 2

A Ferrari 360 Modena (see the figure below) has a
$\dot{W}=410\,{\rm hp}$ engine and mass of $m=1450\,{\rm kg}$.  The
force $F_e$ that propels the car forward is provided by the engine,
and the magnitude of that force is directly related to the power it
supplies.  Recall that when a force acts over a distance, it performs
work.  That is, if a force ${\bf F}$ acts over a small displacement
$d{\bf x}$, then the work performed $dW$ is
\begin{equation*}
\label{eq:work}
dW = {\bf F}\cdot d{\bf x}
.
\end{equation*}
Remember that power is the rate at which work is performed; that is,
it is the work per unit time.  By dividing both sides of
the above equation by the interval of time $dt$ over which the
displacement occurred, we find:
\begin{equation*}
\frac{dW}{dt} = {\bf F}\cdot \frac{d{\bf x}}{dt}
.
\end{equation*}
Using the fact that displacement per unit time is just the velocity,
we find:
\begin{equation*}
\label{eq:power}
\dot{W} = {\bf F}\cdot{\bf v}
.
\end{equation*}
So, power (or the rate at which work is performed) is equal to the
force times velocity.

If the engine outputs a power $\dot{W}$, then the force $F_e$ it
supplies is:
\begin{equation*}
F_e = \frac{\dot{W}}{v}
\end{equation*}
where $v$ is the speed of the car.  

At low speeds, one of the main sources of friction acting on a car is
rolling friction, which acting in a direction opposing the motion of
the car.  The magnitude of this friction $F_r$ is given by:
\begin{equation*}
F_r = -\mu_r m g
\end{equation*}
where $\mu_r$ is the coefficient of rolling friction.  The value of
$\mu_r$ depends on the surface the car is traveling over; for a car on
asphalt, $\mu_r\approx0.03$.  

At higher speeds, air resistance becomes increasingly important.  The
drag force $F_d$ on a moving object due to air resistance is described
by
\begin{align*}
  F_d &= C_d A \frac{1}{2}\rho v^2
%  C_d &= \frac{F_d/A}{\rho v^2/2}
\end{align*}
where $C_d$ is the drag coefficient (a dimensionless quantity
dependent on the shape of the object), $A$ is a reference area
(typically the projected area of the object), and $\rho$ is the
density of the fluid.  The body of the car has a constant drag
coefficient $C_d = 0.34$, based on a frontal reference area of
$A=2.03\,{\rm m}^2$.  Assume the density of air is
$1.225\,{\rm kg\,m}^{-3}$.

## Model solutions for Q2

### Part 1

To estimate the top speed of the car, we multiply the force that
  the engine must exert by the maximum velocity $v$.  This must equal
  to maximum power output of the engine $\dot{W}$.  Note that this
  force will equal the sum of the drag and rolling friction.
\begin{align*}
\dot{W} &= F v 
\\
&= (F_r + F_d) v
\\
&= \left[\mu_r mg + (C_d A) \frac{\rho v^2}{2}\right] v
\end{align*}
Note that this is a cubic equation; while it can be solved
analytically, it is easier to solve it numerically.  Therefore, we
find $v=88.5$\,m\,s$^{-1}$ or $198$\,mph.

```python
from numpy import *
from scipy.optimize import fsolve


def res(v):
    f = (mur*m*g + Cd*A*rho*v**2/2.0)*v - W
    return f



m = 1450.0         # mass of car
Cd = 0.34
A = 2.03           # area of car / m^2
W = 745.7 * 410.0  # engine power / W
rho = 1.225        # density of air / kg m^{-3}
g = 9.81           # acceleration due to gravity / m s^{-2}
mur = 0.03         # coefficient of rolling friction


vmax = fsolve(res, 0.0)
print(vmax, vmax * 2.23694)

```

### Part 2

Substituting our expressions for the forces supplied by the engine and
the rolling friction, we find
\begin{align*}
\frac{d}{dt}(mv) 
&= \frac{\dot{W}}{v} - \mu_r m g - \frac{1}{2}C_d A\rho v^2
\\
dt &= \frac{m}{\dot{W}/v - \mu_r m g  - \frac{1}{2}C_d A\rho v^2}  dv
\\
&= \frac{m v/\dot{W}}{1 - \mu_r m g v/\dot{W} - \frac{1}{2}C_d A\rho v^3/\dot{W}}  dv
.
\end{align*}
If we assume that the car is stationary at time $t=0$, then we can
integrate both sides of the equation to give a the time at which the
car will achieve a particular speed $v$:
\begin{align*}
\int_0^t dt 
&= \int_0^v 
\frac{m v/\dot{W}dv}{1 - \mu_r m g v/\dot{W} - \frac{1}{2}C_d A\rho v^3/\dot{W}}  
.
\end{align*}
In principle, this integral could be performed analytically by determining the roots of the cubic polynomial in the denominator and using partial fractions; however, this would be quite difficult and prone to error.  

Instead, it is much simpler to evaluate this integral numerically using the trapezoid rule.

```python
v_f = 60.0 / 2.23694   # final speed / m s^{-1}


def I_f(v):
    f = m*v/W/(1.0 - (mur*m*g+Cd*A*rho*v**2/2.0)*v/W)
    return f


N = 100+1
v_data = linspace(0.0, v_f, N)

I = 0.0
for k in range(N-1):
    va = v_data[k]
    vb = v_data[k+1]
    I += 0.5*(vb-va)*(I_f(va)+I_f(vb))

t = I
#print(I)

print(f'time = {t} s')

```

We find $t=1.77\,{\rm s}$, which is about a factor of two
shorter than the actual time of $3.7\,{\rm s}$.  This is most likely due to
the fact that we are neglecting the time it takes the engine to reach
full power (as well as the variation of power with speed).

Note that we could have also used the `quad` function from scipy.integrate to perform the integral.

```python
from scipy.integrate import quad

I = quad(I_f, 0.0, v_f)
t = I[0]
#print(I)

print(f'time = {t} s')

```

## Model solutions for Q3

### Part 1

First we complete the square in the argument of the square root:
\begin{align*}
  7+6x-x^2
  &= 16-9+6x-x^2 = 16 - (x-3)^2
\end{align*}
This suggests the substitution:
\begin{align*}
  x &= 3 + 4 \sin\theta
  \\
  dx &= 4\cos\theta d\theta
\end{align*}
We then find
\begin{align*}
  \int \frac{dx}{\sqrt{7+6x-x^2}}
  &= \int \frac{4\cos\theta d\theta}{\sqrt{16 - 16\sin^2\theta}}
  \\
  &= \int \frac{\cos\theta d\theta}{\sqrt{1-\sin^2\theta}}
    = \int d\theta
  \\
  &= \theta + C
  \\
  &= \arcsin\frac{x-3}{4} + C
\end{align*}


```python
import sympy as sp

x = sp.symbols('x')

I = sp.integrate(1/sp.sqrt(7+6*x-x**2), x)
print(I)
```

If we just complete the square of the argument in the square root, we see that this is enough help for `sympy` to figure out how to do the integral.

```python
import sympy as sp

x = sp.symbols('x')

I = sp.integrate(1/sp.sqrt(16 - (x-3)**2), x)
print(I)

```

### Part 3

First we complete the square in the argument of the square root:
\begin{align*}
  4x^2-12x
  &= 4(x^2-3x) = 4(x^2-3x+ 9/4 - 9/4)
    = 4[(x-3/2)^2-9/4]
\end{align*}
This suggests the substitution:
\begin{align*}
  x &= \frac{3}{2} + \frac{3}{2\cos\theta} 
  \\
  dx &= \frac{3}{2\cos^2\theta} \sin\theta d\theta
\end{align*}
We then find
\begin{align*}
  \int dx\,\frac{3x-5}{\sqrt{4x^2-12x}}
  &= \int \frac{3\sin\theta d\theta}{2\cos^2\theta} \,
    \frac{3x-5}{\sqrt{4[9/(4\cos^2\theta)-9/4]}}
  \\
  &= \frac{1}{2}\int \frac{\sin\theta d\theta}{\cos^2\theta} \,
    \frac{(3x-5)\cos\theta}{\sqrt{1-\cos^2\theta}}
  \\
  &= \frac{1}{2}\int d\theta\frac{3x-5}{\cos\theta}
  \\
  &= \frac{1}{2}\int d\theta\frac{9(1+1/\cos\theta)/2-5}{\cos\theta}
  \\
  &= \frac{1}{4}\int d\theta
    \left[\frac{9}{\cos^2\theta}- \frac{1}{\cos\theta}\right]
  \\
  &= \frac{1}{4}\int d\theta
    \left[\frac{9}{\cos^2\theta}- \frac{\cos\theta}{\cos^2\theta}\right]
  \\
  &= \frac{1}{4}\int d\theta
    \left[\frac{9}{\cos^2\theta}- \frac{\cos\theta}{1-\sin^2\theta}\right]
  \\
  &= \frac{1}{4}\int d\theta
    \left[\frac{9}{\cos^2\theta}
    - \frac{\cos\theta}{2(1+\sin\theta)}
        - \frac{\cos\theta}{2(1-\sin\theta)}
    \right]
  \\
  &= \frac{1}{4}
    \left[9\tan\theta - \frac{1}{2}\ln\frac{1+\sin\theta}{1-\sin\theta}\right]
    + C'
  \\
  &= \frac{1}{4}
    \left[9\tan\theta - \frac{1}{2}\ln\frac{(1+\sin\theta)^2}{\cos^2\theta}\right]
    + C'
  \\
  &= \frac{1}{4}
    \left[9\tan\theta - \ln(1/\cos\theta+\tan\theta)\right]
    + C'
  \\
  &= \frac{1}{4}
    \left[9\sqrt{(2x/3-1)^2-1}
    - \ln((2x/3-1)+\sqrt{(2x/3-1)^2-1})\right]
    + C'
  \\
  &= \frac{3}{4}\sqrt{4x^2-12x}
    - \frac{1}{4}\ln(2x-3+\sqrt{4x^2-12x})
    + C
\end{align*}


```python
import sympy as sp

x = sp.symbols('x')

I = sp.integrate((3*x-5)/sp.sqrt(4*x**2-12*x), x)
print(I)
```

This integral can be performed by repeatedly applying integration by
parts:
\begin{align*}
x^2 e^{-x}
&= \frac{d}{dx}[-x^2 e^{-x}]
+ 2 x e^{-x}
\\
&= \frac{d}{dx}[-x^2 e^{-x}]
+ \frac{d}{dx}[- 2 x e^{-x}]
+ 2 e^{-x}
\\
&= \frac{d}{dx}[-x^2 e^{-x}]
+ \frac{d}{dx}[- 2 x e^{-x}]
+ \frac{d}{dx}[- 2 e^{-x}]
\\
&= \frac{d}{dx}[-x^2 e^{-x} - 2 x e^{-x} - 2 e^{-x}]
\end{align*}
Therefore, we find
\begin{align*}
\int dx\,x^2 e^{-x}
&= \int dx\,\frac{d}{dx}[-x^2 e^{-x} - 2 x e^{-x} - 2 e^{-x}]
\\
&= -(x^2 + 2 x + 2) e^{-x} + C
\end{align*}
